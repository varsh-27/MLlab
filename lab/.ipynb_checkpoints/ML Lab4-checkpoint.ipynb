## Housing Case Study
Problem Statement:

Consider a real estate company that has a dataset containing the prices of properties in the Delhi region. It wishes to use the data to optimise the sale prices of the properties based on important factors such as area, bedrooms, parking, etc.

Essentially, the company wants —

To identify the variables affecting house prices, e.g. area, number of rooms, bathrooms, etc.

To create a linear model that quantitatively relates house prices with variables such as number of rooms, area, number of bathrooms, etc.

To know the accuracy of the model, i.e. how well these variables can predict house prices.

```python
#1 Import the dataset (Housing.csv) and look at the top five rows
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/home/rgukt-basar/Downloads/ML Lab/datasets/Housing.csv')
df
```

```python
#2 Check the shape of the DataFrame (rows, columns)
df.shape
```

```python
#3 Look at the data types of the columns
df.info()
```

```python
#4 Check for missing  values if any, replace with appropriate values
df.isnull().sum()
```

```python
#5 Prepare  X (independent variables) and y (dependent variable)
x=df.drop(columns=['price'])
y=df['price']
```

```python
#6 Visualise the relationship between the independent variables and the depenedent variable using scatterplots
sns.pairplot(df)
plt.show()
```

```python
#7 Encoding categorical data in X
categorical_cols=df.select_dtypes(include=['object','category']).columns
label=LabelEncoder()
for col in categorical_cols:
    df[col]=label.fit_transform(df[col])
df
```

```python
#8  Avoiding the Dummy Variable Trap
df1=pd.DataFrame({'furnishingstatus':['furnished','semifurnished','unfurnished']})
df1['furnishingstatus']=df1['furnishingstatus'].str.strip()

encoded=pd.get_dummies(df1,columns=['furnishingstatus'],drop_first=True)
print(encoded)

```

```python
#9 Apply feature scaling on numerical variables
scaler=StandardScaler()
num_cols=df.select_dtypes(include=['number']).columns
for col in num_cols:
    df[num_cols]=scaler.fit_transform(df[num_cols])
df  
```

```python
#10 Split Data into Training and Testing Sets (70%-Train 30%-Test) - X_train, y_train, X_test, y_test
x=df.drop(columns=['price'])
y=df['price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
x_train.shape
```

```python
#11 Fitting Multiple Linear Regression to the Training
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#12 Predict on the train set and calculate the error = y_pred - y_train
y_pred=model.predict(x_train)
error=y_pred-y_train
error
```

```python
#13  Residual plot - plot y_pred in x axis and errors in y axis
plt.scatter(y_pred,y_train,alpha=0.2)
plt.axhline(y=0, color='black', linestyle='--') 
plt.xlabel("y_pred")
plt.ylabel("y_train")
plt.grid(True)
plt.show()
```

```python
#14  Predic on the test set
y_test_pred=model.predict(x_test)
```

```python
#15 Residual plot - plot y_pred in x axis and errors in y axis
plt.scatter(y_test_pred,y_test,alpha=0.2)
plt.axhline(y=0, color='black', linestyle='--') 
plt.xlabel("y_test_pred")
plt.ylabel("y_test")
plt.grid(True)
plt.show()
```

```python
#16 Print Mean Squared Error and R Squared Value
msv=mean_squared_error(y_test,y_test_pred)
rsv=r2_score(y_test,y_test_pred)
print("Mean squared value : ",msv)
print("r-squared value : ",rsv)
```
