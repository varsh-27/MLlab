```python
#Use bernoullNB for training
from sklearn.naive_bayes import BernoulliNB
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer to transform text data into numerical features
vectorizer = CountVectorizer()

# Fit the vectorizer to the training data and transform it
X_train = vectorizer.fit_transform(twenty_train.data)

# Create a BernoulliNB classifier
bernoul = BernoulliNB()

# Fit the classifier to the transformed training data and target labels
bernoul.fit(X_train, twenty_train.target) # Use X_train instead of twenty_train
```

```python
#Use bernoullNB for training
from sklearn.naive_bayes import BernoulliNB
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer to transform text data into numerical features
vectorizer = CountVectorizer()

# Fit the vectorizer to the training data and transform it
X_test = vectorizer.fit_transform(twenty_test.data)

# Create a BernoulliNB classifier
bernoul = BernoulliNB()

# Fit the classifier to the transformed training data and target labels
bernoul.fit(X_test, twenty_test.target) # Use X_train instead of twenty_train
```

```python
#predict target labels for testing set
y_pred=bernoul.predict(X_test)
y_pred
```

```python
#find accuracy score on test set
from sklearn.metrics import accuracy_score
accuracy_score(twenty_test.target,y_pred)
```
