```python
#replacing . values with nan
df["4-year resale value"]=df["4-year resale value"].replace('.',np.nan)
df["Price in thousands"]=df["Price in thousands"].replace('.',np.nan)
df["Engine size"]=df["Engine size"].replace('.',np.nan)
df["Horsepower"]=df["Horsepower"].replace('.',np.nan)
df["Width"]=df["Width"].replace('.',np.nan)
df["Length"]=df["Length"].replace('.',np.nan)
df["Curb weight"]=df["Curb weight"].replace('.',np.nan)
df["Fuel capacity"]=df["Fuel capacity"].replace('.',np.nan)
df["Fuel efficiency"]=df["Fuel efficiency"].replace('.',np.nan)
df["Sales in thousands"]=df["Sales in thousands"].replace('.',np.nan)
df["Wheelbase"]=df["Wheelbase"].replace('.',np.nan)
```

```python
#changing datatypes to float
df["4-year resale value"]=df["4-year resale value"].astype('float64')
df["Sales in thousands"]=df["Sales in thousands"].astype('float64')
df["Price in thousands"]=df["Price in thousands"].astype('float64')
df["Curb weight"]=df["Curb weight"].astype('float64')
df["Engine size"]=df["Engine size"].astype('float64')
df["Horsepower"]=df["Horsepower"].astype('float64')
df["Width"]=df["Width"].astype('float64')
df["Length"]=df["Length"].astype('float64')
df["Fuel capacity"]=df["Fuel capacity"].astype('float64')
df["Fuel efficiency"]=df["Fuel efficiency"].astype('float64')
df["Wheelbase"]=df["Wheelbase"].astype('float64')
```

```python
#Print number of missing values in each column
df.isnull().sum()
```

```python
#Drop the column with most missing values
maxMissing=df.isnull().sum().idxmax()
df=df.drop(columns=[maxMissing])
df
```

```python
# Drop the rows with categorical missing values
category=df.select_dtypes(include=['object','category']).columns
df=df.dropna(subset=category)
df.shape
```

```python
#Import the rows with numerical missing values
numerical_cols = df.select_dtypes(include=['number']).columns
rows_with_numerical_na = df[df[numerical_cols].isnull().any(axis=1)]
rows_with_numerical_na
```

```python
#Replacing missing values with mean
#df["4-year resale value"].fillna(df['4-year resale value'].mean(),inplace=True)
df["Sales in thousands"].fillna(df['Sales in thousands'].mean(),inplace=True)
df["Price in thousands"].fillna(df['Price in thousands'].mean(),inplace=True)
df["Curb weight"].fillna(df['Curb weight'].mean(),inplace=True)
df["Engine size"].fillna(df['Engine size'].mean(),inplace=True)
df["Horsepower"].fillna(df['Horsepower'].mean(),inplace=True)
df["Width"].fillna(df['Width'].mean(),inplace=True)
df["Length"].fillna(df['Length'].mean(),inplace=True)
df["Fuel capacity"].fillna(df['Fuel capacity'].mean(),inplace=True)
df["Fuel efficiency"].fillna(df['Fuel efficiency'].mean(),inplace=True)
df["Wheelbase"].fillna(df['Wheelbase'].mean(),inplace=True)
```

```python
#Sort the data w.r.t price_find the details of the most and the least expensive cars
df=df.sort_values(by=['Price in thousands'],ascending=True)
print("Least expensive : \n",df.iloc[0])
print("Most Expensive : \n",df.iloc[-1])
```

```python
#Write a function to find min and max values of any column
def get_min_max(df, column_name):
    if column_name in df.columns:
        min_val = df[column_name].min()
        max_val = df[column_name].max()
        return min_val, max_val
```

```python
#Call the above function to find min max of horse power, length, fuel efficiency
min_val,max_val=get_min_max(df,'Horsepower')
print("Minimum value of Horsepower : ",min_val)
print("Maximum value of Horsepower : ",max_val)
min_val1,max_val1=get_min_max(df,'Fuel efficiency')
print("Minimum value of Fuel efficiency : ",min_val1)
print("Maximum value of Fuel efficiency : ",max_val1)
```

```python
#Plot histogram of continuous numerical variable : price,sales,hoursepower,fuelefficiency
def plot_histograms(df, columns):
    for col in columns:
        if col in df.columns:
            plt.figure(figsize=(6, 4))
            plt.hist(df[col].dropna(), bins=30, color='skyblue', edgecolor='black')
            plt.title(f'Histogram of {col}')
            plt.xlabel(col)
            plt.ylabel('Frequency')
            plt.grid(True)
            plt.show()
        else:
            print(f"Column '{col}' not found in DataFrame.")
plot_histograms(df, ['Price in thousands', 'Sales in thousands', 'Horsepower', 'Fuel efficiency'])
```
