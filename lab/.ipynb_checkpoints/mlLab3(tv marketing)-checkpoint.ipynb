```python
#Visualize the relationship between the features and the dependent variable using scatterplots
plt.figure(figsize=(8,6))
plt.scatter(df['TV'],df['Sales'])
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Relationship between the features and the dependent variable')
plt.grid(True)
plt.show()
```

```python
#Prepare X (feature variables) and y (response variable)
x=df['TV']
y=df['Sales']
```

```python
#Split Data into Training and Testing Sets (70%-Train 30%-Test) - X_train, y_train, X_test, y_test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
#Print the shape of train data and test data
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Perform Linear Regression (on train data) without using any library and print the coefficients values

#def predict(x):
#return a+b*x
x = x_train.to_numpy().flatten()
y = y_train.to_numpy().flatten()

n = len(x)
x_mean = sum(x) / n
y_mean = sum(y) / n

numerator = sum((x[i] - x_mean) * (y[i] - y_mean) for i in range(n))
denominator = sum((x[i] - x_mean) ** 2 for i in range(n))

b = numerator / denominator
a = y_mean - b * x_mean

print("Slope (b):", b)
print("Intercept (a):", a)
```

```python
#.Create the model using the calculated coefficients and make predictions on the testing set
def predict(x):
    return a+b*x
y_pred=predict(x_test)
y_pred
```

```python
#.Perform Linear Regression using library(sklearn)
# Reshape if you have a single feature
x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
model=LinearRegression()
model.fit(x_train,y_train)
a = model.intercept_  # Intercept
b = model.coef_[0] 
```

```python
#Print the intercept and coefficients
print('intercept',a)
print('slope',b)
```

```python
#Make predictions on the testing set - y_pred
x_test=x_test.reshape(-1,1)
y_pred=model.predict(x_test)
```

```python
#Plot test data and the model
y_train_pred = model.predict(x_train)
x_vals = x_train.flatten()
y_vals = y_train.flatten()
y_pred_vals = y_train_pred.flatten()
plt.scatter(x_vals, y_vals, color='blue', label='Train Data', alpha=0.6)
plt.plot(x_vals, y_pred_vals, color='red', linewidth=2, label='Regression Line')
plt.title("Linear Regression - Train Data vs Model")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#Plot test data and the model
y_test_pred=model.predict(x_test)
x1=x_test
y1=y_test
y_pred=y_test_pred
plt.scatter(x1,y1,color='green',label='Test data')
plt.plot(x1,y_pred,color='red',label='Regression Line')
plt.title("Linear Regression - Test Data vs Model")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()
```

```python
#Print Mean Squared Error and R Squared Value
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y1, y_pred)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

```python
#Calculate and print R Squared and Adjusted R Squared values without any library
import numpy as np

# Make predictions
y_pred = model.predict(x_train)

# Flatten arrays
y_true = y_train.flatten()
y_pred = y_pred.flatten()

# Number of samples and features
n = len(y_true)
k = x_train.shape[1]  # number of features

# Calculate SS_res and SS_tot
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

# R Squared
r_squared = 1 - (ss_res / ss_tot)

# Adjusted R Squared
adj_r_squared = 1 - ((1 - r_squared) * (n - 1) / (n - k - 1))

# Print results
print("R Squared (R²):", round(r_squared, 4))
print("Adjusted R Squared:", round(adj_r_squared, 4))
```
